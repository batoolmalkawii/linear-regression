import packages
___

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

____
#### Read dataset
___

```python
dataset = pd.read_csv('./Fish.csv')
dataset = dataset.dropna()
```

___
Show the first five rows of the data set.
___

```python
dataset.head(5)
```

___
Split dataset
___


```python
X = dataset['Width'].values
Y = dataset['Weight'].values
X = X.reshape(-1,1)
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.7, test_size=0.3, random_state=100)
```

___
#### Visualize dataset
___

```python
%matplotlib inline
plt.scatter(x_train, y_train, color='purple')
plt.show()
```

___
#### Show description and info
___

```python
description = dataset.describe()
info = dataset.info()
print(description)
print(info)
```

___
#### Linear Regression model
___

```python
model = LinearRegression()
model.fit(x_train, y_train)
y_predict = model.predict(x_test)
```

___
#### Model Accuracy
___

```python
print(f'Training Accuracy: {model.score(x_train, y_train)*100}%')
print(f'Testing Accuracy: {model.score(x_test, y_test)*100}%')
```

___
#### Conclusion #1
___

```python
%matplotlib inline
plt.scatter(x_train, y_train, color='purple')
plt.scatter(x_test, y_test, color='y')
plt.plot(x_test, y_predict, color='pink')
plt.ylabel('Weight')
plt.xlabel('Width')
plt.show()
```

```python
X = dataset['Height'].values
Y = dataset['Weight'].values
X = X.reshape(-1,1)
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.7, test_size=0.3, random_state=100)
model = LinearRegression()
model.fit(x_train, y_train)
y_predict = model.predict(x_test)
plt.scatter(x_train, y_train, color='black')
plt.plot(x_test, y_predict, color='yellow')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```
